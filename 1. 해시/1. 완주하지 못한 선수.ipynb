```python
# 효율성(시간 초과)때문에 오답

def solution(participant, completion):
    for person in completion:
        if person in participant:
            participant.remove(person) # pop(index), remove(value)
            continue
            
    answer = participant.pop()
    return answer
```

```python
# 정답!

def solution(participant, completion):
    participant.sort()
    completion.sort()
    for p, c in zip(participant, completion):
        if p != c:
            return p
    
    return participant[-1]  # participant가 completion보다 요소가 1개 더 많으므로, 맨 마지막 전까지 전부 p == c 일 경우! (마지막이 답!)
```

```python
# 다른 풀이 (collections.Counter().keys() 이용)

import collections

def solution(participant, completion):
    answer = collections.Counter(participant) - collections.Counter(completion)
    return list(answer.keys())[0]
```

## [ collections 모듈 - Counter() ]
컨테이너에 **동일한 값의 요소가 몇 개씩인지**를 파악하는데 사용하는 **객체**이다.  
* import collections 해서 사용
* 결과값(return)은 **딕셔너리 형태**로 출력된다. ==> **('key' : 'value')**
    >* 따라서 **.keys()** 함수로 중복을 제외한 요소 종류인 키 값만 뽑아낼 수 있다. (**values()** 도 가능)  
    >* **list()** 함수까지 사용하면, 리스트 형태로 키 값 종류들 뽑아낼 수 있음. (중복 제외)
* collections.Counter()는 **산술/집합 연산**이 가능하다.
    > ex. collections.Counter(participant) - collections.Counter(completion)

```python
import collections

c = collections.Counter('aaabbcdddd') # c (컨테이너) : Counter({'d': 4, 'a': 3, 'b': 2, 'c': 1})

participant = ['leo', 'kiki', 'eden']
completion = ['eden', 'kiki']

par_c = collections.Counter(participant)
com_c = collections.Counter(completion)

print('collections.Counter(participant) :', par_c)
print('collections.Counter(completion) :', com_c)
print()
print('par_c - com_c :', par_c - com_c)
print('par_c + com_c :', par_c + com_c)
print()
print('par_c.keys() :', par_c.keys())
print()
print('list(par_c.keys()) :', list(par_c.keys()))
```
